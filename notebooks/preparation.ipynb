```python
# Importing libraries
import pandas as pd
import json
```

```python
# Read JSON data from file
file_path = "results_user_eval_RAW.json"  # Adjust the file path if needed
with open(file_path, 'r') as file:
    json_data = json.load(file)

# Creating a Pandas DataFrame
df = pd.DataFrame(json_data)
```

```python
df.head()
```

```python
# Save DataFrame to CSV
csv_file_path = "result_user_eval_RAW.csv"
df.to_csv(csv_file_path, index=False)
```

```python
# List of columns to delete
columns_to_delete = [
    "Accept", "Accept-Encoding", "Accept-Language", "Host", "Referer", 
    "Sec-Ch-Ua-Mobile", "Sec-Fetch-Dest", "Sec-Fetch-Mode", "Sec-Fetch-Site", 
    "Sec-Fetch-User", "Upgrade-Insecure-Requests", "X-Forwarded-For", 
    "X-Forwarded-Host", "X-Forwarded-Port", "X-Forwarded-Proto", 
    "X-Forwarded-Server", "X-Real-Ip", "t0_end", "t0_start", "t1_end", 
    "t1_start", "Dnt", "Sec-Gpc", "Te"
]

# Drop the specified columns
df.drop(columns=columns_to_delete, inplace=True)
df.head()
```

```python
# Save Cleaned DataFrame to CSV
csv_file_path = "result_user_eval_CLEAN.csv"
df.to_csv(csv_file_path, index=False)
```
